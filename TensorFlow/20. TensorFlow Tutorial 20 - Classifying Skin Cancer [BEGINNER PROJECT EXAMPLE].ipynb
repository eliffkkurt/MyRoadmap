### <font color="blue"> Giriş</font>

#### Problem

Deri dokularının resimlerini sınıflandıran bir network kurmaya çalışacağız benign (not harmful) ve malignant(cancer)ı ifade ediyor.

#### ---

### Dataset

[Isic Challenge](https://challenge.isic-archive.com/data) sitesinden 2019 datasetini indiriyoruz. 

* Sadece training datasetiyle işlem yaptık.

* Training data setinin groundtruth.csv dosyasına baktığımızda resimlerin dosya adını, malignant olup olmadığı bilgisini ve deri dokusunun türü hakkında bilgi veren sütunları görüyoruz. Bu notebookta biz deri dokusunun türüyle ilgilenmeyeceğiz. Sadece kanser olup olmadığıyla ilgileneceğiz. 


* Training data setinin metadata.csv dosyasına baktığımızda resimlerin dosya adı bilgisini, ilgili kişinin yaşı, dokunun nerede konumlandığını ve cinsiyet bilgisini görüyoruz. Bu notebookta bu dosyayla hiç ilgilenmeyeceğiz. Fakat modelimizi geliştirmek istiyorsak yaş bilgisi önemli olacaktır burada.


* **Dosya düzeni**: "ISIC" adında bir klasör açtık. Onun içine indirdiğimiz tüm klasörleri koyduk. İndirdiğimiz klasörlerin adlarını *ISIC_2019_Training_Input -> images*, *ISIC_2019_Training_Metadata -> additional_info*, *ISIC_2019_Training_GroundTruth -> labels* olarak değiştirdik. Bundan sonraki klasör düzenleme işlemlerini kod ile yaptık.

1. Train, validation ve test setimizi oluşturduk kod ile.

2. Her bir klasöre kaç resim koyacağımızı ayarlıyoruz ve resimleri bu sayıya göre başka klasörlere kopyalıyoruz.
