# Reklam Harcaması Optimizasyonu

Bu bölümde binom dağılımını kullanarak bir iş problemini çözmeyi öğreneceğiz.

**Problem:**

Çeşitli mecralara reklam veriliyor, reklamların tıklanma ve geri dönüşüm oranları optimize edilmeye çalışılıyor. Buna yönelik olarak belirli bir mecrada çeşitli senaryolara göre reklama tıklama olasılıkları hesaplanmak isteniliyor.

Öyle bir plan yapmam lazım ki optimum bir şekilde bütçemi kullanıp maksimum bir şekilde geri dönüş alayım. Bu noktada bazı olasılıksal değerler bize bilinmezliği modellemek anlamında değerli bilgiler sunabilmektedir. 

**Detaylar:**

- Bir mecrada reklam verilecek.

- Dağılım ve reklama tıklama olasılığı biliniyor (0.01). Yani daha önce bu işlemler yapılmış. Şirket veri bilimi kültürü olan bir şirket. Buna benzer ölçümleri yapmış ve şunu biliyor: reklama tıklanma olasılığımız 0.01.

**Soru:**

<font color='blue'> Reklamı 100 kişi gördüğünde 1, 5 ya da 10 kişinin tıklaması olasılığı nedir? </font>

Problem ne? Problem şu: ben bir platforma reklam vereceğim fakat bunun dönüşüyle ilgili bir şey bilmiyorum. Bunun için olasılık hesaplayabilirim. Buna göre kararlarımı almak üzere bir veri olur elimde, reklam harcamalarımı optimize edebilirim. 

Yani örneğin 100 kişiyi göz önünde bulundurduk. Demiş olacağız ki bu reklamın olasılığı biliniyor, binom dağılımı olduğu biliniyor. Gerçekleşen deneylerin yani bu reklamı gören kişilerin buna tıklama ya da tıklamama olaylarının eşit şartlar altında gerçekleştiği varsayılıyor. 

Bunlara göre olasılık hesaplama işlemlerini gerçekleştirelim.

<font color='blue'>  

* n = 100 
    
* Kaç başarı bekliyoruz: 1
    
* 1 tıklanması olasılığını inceliyoruz. 
    
</font>
Değerleri yerine yazdığımızda bu reklamı 100 kişi gördüğünde 1 kişinin tıklaması olasılığı 0.37. Oldukça yüksek bir oran reklam sektörüne göre.

### Binom Dağılımı

${\displaystyle f(k;n,p)={n \choose k}p^{k}(1-p)^{n-k}}$, k = 0, 1, 2, ...,n


${\displaystyle \operatorname {E} (X)=np}$


${\displaystyle \operatorname {Var} (X)=np(1-p)}$

```python
from scipy.stats import binom
```

```python
n = 100
p = 0.01
rv = binom(n,p)

#pmf: olasılık kütle fonksiyonu
print(rv.pmf(1))  # 1 kişinin tıklaması
print(rv.pmf(5))  # 5 kişinin tıklaması
print(rv.pmf(10)) # 10 kişinin tıklaması
```

Amacımız belirsizlikle ilgili belirli bir şeylere ulaşmak. Elimizde hiçbir değer olmamasındansa bu çeşitli mecralarla ilgili bu şekilde olasılıksal bazı değerlerin olması her zaman tercih edecek olduğumuz durumdur. Bu tip olasılık dağılımları bize, veriye dayalı hareket etme noktasında bunları nümerik olarak da ölçüp onlara göre kararlar alma imkanı sağlar. Örneğin bir reklamı 100 kişi gördüğünde 5 kişinin tıklama olasılığı ile 10 kişinin tıklama olasılığı karşılaştırıldığında bu reklam için google'a veya herhangi bir platforma ne kadar yatırım yapılması gerektiği konusunda fikir sahibi oluruz. Mesela arada çok fark yoksa 10 kişinin tıklamasına yönelik olarak **10 kişinin tıklaması için bu reklamı kaç kişi görmeli** sorusunu sorarız.
