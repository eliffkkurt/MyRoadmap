## İş Uygulaması: İlan Girişi Hata Olasılıklarının Hesaplanması

Bir ilan sitesi var. Bu ilan sitesinde bazı ilan girişleri yapılıyor. El yordamıyla olabilir, makine yordamıyla yapılıyor olabilir. Bunlara ilişkin bir hata hesaplanmak isteniyor. 

Bir bilinmezlik var. Örneğin ben bu hataların olasılığını bilip belki dönemlik eleman istihdam edeceğim. Belki bir sistem yenilemesi gerçekleştireceğim. Fakat belirli dönemler için ne kadar hata olabileceğini, hata olasılıklarını bilmiyorum (bilinmezlik var). İşte bu bilinmezlik durumunu görece bazı yöntemlerle ortadan kaldırmaya çalıştığımızda Poisson dağılımı devreye girip şunu demiş olacak: hatalı ilan girişi hiç olmaması olasılığı bu, 3 olması ve 5 olması olasılığı bu demiş olacak. 

Bunu biz bir iş yeri olarak elemanlarımıza hedef vermek, ek eleman talebinde bulunmak vb. ihtiyaçlarla kullanabiliriz. Hareketlerimizi, davranışlarımızı bilimsel bir forma oturtturabiliriz. Buna karşılık diyebiliriz ki biz kararlarımızı veriye dayalı olarak alıyoruz. Örneğin şunu demiş olacak ilgili kişi: biz yaz döneminin şu döneminde hata olasılıklarımızı incelediğimizde bunların dağılımını biliyoruz Poisson ve bunun neticesinde de hata olasılıklarımızın sayısı artmakta, olasılık değerlerinin arttığını gözlemliyoruz. İşte bu sebeple eleman alımında bulunacağız.

**Problem:** 

- Hatalı ilan girişi olasılıkları hesaplanmak isteniyor.

**Detaylar:** 
- Bir yıl süresince ölçümler yapılıyor.
- Dağılım biliniyor (Poisson) ve Lambda 0.1 (ortalama hata sayısı)(1 yıl boyunca ölçüm yapıldıysa hatalı işlemler de sağlıklı işlemler de zaten biliniyordur.)
- Hiç hata olmaması, 3 hata olması ve 5 hata olması olasılıkları nedir? 

### Poisson Dağılımı

${\displaystyle f(x,\lambda )={\frac {\lambda ^{x}e^{-\lambda }}{x!}}\,\!}$,  x = 0,1,2,...,n

Hiç hata gerçekleşmemesi olasılığı %90'mış, gayet içgüdüye uygun. Çünkü nadir bir olaydı. %90'ı başarılı gerçekleşiyor bu işlemlerin. 5 hata olma olasılığı 3 hata olma olasılığından daha küçük ki bu da içgüdüye uygun.

```python
from scipy.stats import poisson
```

```python
lambda_ = 0.1 # direkt lambda yazamadım böyle bir fonksiyon varmış.
```

```python
rv = poisson(mu = lambda_) # mu dediğimiz şey lambda 

# bunu çalıştırdığımızda poisson nesnesi oluşmuş oluyor.
```

```python
print(rv.pmf(k=0)) # k: başarı sayısı
```

```python
print(rv.pmf(k=3))
```

```python
print(rv.pmf(k=5))
```
