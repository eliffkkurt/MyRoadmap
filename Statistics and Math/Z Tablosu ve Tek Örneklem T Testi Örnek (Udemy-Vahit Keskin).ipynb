## İş Uygulaması: Ürün Satın Alma Adım Optimizasyonu

**Z Tablosu ve Tek Örneklem T Testi Örnek**

Bir e-ticaret şirketi var, internette bazı ürünler satıyor ve bu ürünleri satması sürecinde 4-5 adım var. Sepete ürünün eklenmesi, kart bilgilerinin girilmesi vs. Bu adımların kullanıcı deneyimi tarafında optimize edilmesi gerekiyor. Yani ana hedef; buradaki deneyimi mümkün olduğu kadar basit, hızlı bir şekilde gerçekleştirmeye çalışmak ve burada ortaya çıkan problemleri çözmeye çalışmak.

**Problem:** 

Sepete ürün ekleme işlemi sonrasında ödeme ekranında 5 adet adım var ve bu adımlardan bir tanesi sorgulanmaktadır. 

**Detaylar:**
- Her adımın 20'şer sn. olması hedefi var ve buna yönelik olarak 4. adım sorgulanıyor.
- Bu durumu test etmek için 100 örnek (gözlem) alınıyor.
- Örneklemin standart sapması 5 saniyedir. Örneklemin ortalaması ise 19 saniyedir.

Ürünümüzü kullanan binlerce belki onbinlerce müşterimiz var. Bu müşterilerin her birisi bu satın alma sürecinde bir zaman geçiriyor ve bu onbinlerce kullanıcının yine tüm kullanıcıları temsil etmediği biliniyor. Yani belki yüz binlerce olacak, daha bizim sistemimize gelmeyen diğer birçok müşteri var. Burada hiçbir şekilde ana kitleye (populasyona) erişemiyoruz yani. Bu sebeple mevcut müşterilerin arasından belirli bir örneklem alınıyor, 100 olabilir 1000 olabilir. Ve bu örneklem üzerinden bir hipotez testi kurup bunun gerçekten 20 sn mi olduğunu yoksa değil mi test etmiş olacağız. 

**Kritik soru şu:** 100 örnek çektik zaten ortalaması 19 neyin testini yapıyoruz? 

Bu anlaşıldığında hipotez testlerinin mantığı anlaşılacak. Biz şu yüzden test yapıyoruz. Ortaya çıkan bir etkinin şans eseri olup olmadığını istatistiki olarak şansa yer bırakmayacak şekilde ispat etmeye çalışıyoruz. Bu örneklemin ortalaması var evet ama standart sapmasını da göz önünde bulundurarak istatistiki olarak anlamlı olup olmadığını değerlendirmiş olacağız. 

Bu hipotezlerin yorumlarının yanına yazılmasını yapacak olduğunuz çalışmalarda tavsiye ederim. Bunların ne ifade ettiği, buradaki matematiksel formülasyonlardan çok daha önemli.  

**Ho=mü=20 ne demek?** Bilinmeyen anakütle parametresi "mü" yani kişilerin 4. adımda geçirdikleri süre ortalaması 20'dir demek. 

**Ho: mü eşit değildir 20 ne demek?** Bilinmeyen anakütle ortalaması yani kişilerin, tüm insanların 4. adımda geçirdikleri süre 20 snden farklıdır demek. 

Anlamlılık düzeyi dediğimiz şey: alfa = 0.05. Yani çalışmanın başında kabul ettiğimiz kabul edilebilir hata miktarımız. Kurmuş olduğumuz hipotez testi iki yönlü olduğundan dolayı çalışmanın başında kabul ettiğimiz alfayı 0.05'i ikiye bölüyoruz. Çünkü kurmuş olduğumuz hipotez iki yanlı. Eğer büyüktür ya da küçüktür şeklinde bir alternatif hipotezimiz olsaydı bu durumda alfayı olduğu gibi kabul edip bu alfaya karşılık gelen tablo değerine bakacaktık. Fakat alternatif hipotezimiz iki yönlü olduğundan dolayı bu sebeple alfayı ikiye bölüp bunun üzerinden bir Z tablo kritik değeri bulacağız. Z tablo olasılık değerini bulmak için 0.5'ten 0.025'i çıkarıp bu 0.475 değerine karşılık gelen tablo değerine bakmamız gerekiyor ki bunun karşılığı da 1.96. 

### <font color='blue'> Z Tablosu</font>

Tablo üzerinden bakabiliriz. Bu uygulamada ana kitlenin standart sapmasını bilmiyorduk fakat n 30'dan büyük olduğundan dolayı z dağılımını kullandık. n yeteri kadar büyük t dağılımına ilişkin tablo da kullanılabilirdi. 

Yukarıda görmüş olduğumuz alanın tamamı 1'e karşılık geliyor. Bu alanın sağ tarafı (0'dan sonrası) 0.5'e, sol tarafı (0'dan öncesi) 0.5'e karşılık geliyor. İşte bizim çalışmanın başında kabul edilebilir hata miktarı olarak belirlemiş olduğumuz alfayı ikiye böldükten sonra elde ettiğimiz değer bu tabloda verilen eğrinin altında kalan alanları ifade eden bu değerler. 0.025 sağ tarafta taranmamış yeri ifade ediyor. Tüm grafiğin yarısından 0.5'ten taranmamış alan olan 0.025'i çıkardığımızda 0.475 buluyoruz. Burada 0.475 değerini bulduktan sonra bu olasılık değerine karşılık gelen Z değerini bulacağız. Bunu y ekseninden ve x ekseninden karşılaştırarak okumaya çalıştığımızda 1.96 değerini gözlemliyoruz. 

Hatırlayacaksınız güven aralığı bölümünde genellikle %95 güvenilirlilik ile çalıştığımızda alfa değeri 1.96 olarak kabul etmemiz gerektiğini ifade etmiştik. Çünkü orada da artı yönlü ve eksi yönlü işlem yapıyor olduğumuzdan dolayı güven aralığı söz konusu olduğunda 1.96'yı z tablo değeri olarak kabul etmiştik. 

Pratikte bunları elle yapmayacağız.

Bu örneğimizde n, 30'dan büyük olduğundan dolayı z'yi kullandık. Burada t de kullanabilirdik. Çok yakın sonuçlar olacaktır ve biz genelde programlarda t'yi kullanıyor olacağız. 

z test istatistiğinde ilgili değerleri yerine yazıyoruz. Örnek üzerinden bulmuş olduğumuz değer 19. Sınamış olduğumuz hipotezdeki teorik değerimiz 20. Çalışmanın standart sapması elimizde olmadığı için bizim hesaplamış olduğumuz standart sapmayı kullanıyoruz. Yani ana kitleye ilişkin standart sapma değeri elimizde yok. Örneği ifade eden s: standart sapma değeri elimizde. Biz kendimiz hesapladık bunu. Bunu da formülasyonda yerine yazdığımızda z hesap istatistiğimiz -2 olarak belirlendi. 

Elimizde z tablo değeri var ve bir z hesap değeri var. Şimdi bu ikisini karşılaştırarak karar vereceğiz. 

İki yönlü bulduğumuzdan dolayı Zhesap eksili çıktğından dolayı Ztabloyu eksili bir şekilde değerlendiriyoruz yani -1.96 şeklinde değerlendiriyoruz. Zhesap küçükse tablo değerinin eksilisinden veya Zhesap büyükse artılısından H0'ı reddetmiş olacağız. 

Burada görülen yeşil ifadeler karar noktalarını ifade ediyor. Bizim tablomuz üzerinden bulmuş olduğumuz değer 1.96, -1.96 hesaplamış olduğumuz değer ise Zhesap=-2 değeridir. -2 değeri red bölgesi olan H0 hipotezinin red bölgesi olan yeşil bölgeye düşmektedir. Bu sebeple H0 hipotezi reddedililir şeklinde yorumluyoruz. Eğer elimizde pozitif bir hesap değeri olsaydı bu durumda bu grafiğin sağ tarafındaki yeşil alan ile karşılaştırıp +1.96'yla karşılaştırmış olacaktık ve buna göre yorum yapacaktık. 

Eğer hesaplamış olduğumuz z hesap değeri buradaki yeşil alanlara yani hipotez red bölgelerine düşmeseydi ne olacaktı? Bu durumda H0 hipotezini reddetmek için yeterli kanıt bulunamamıştır diyecektik. İstatistiki olarak anlamlı bir sonuç elde edilememiştir diyecektik. 

H1'e atıfta bulunarak yorum yaptık çünkü H0'ı reddettik. Demek ki ana kitle dediğimiz yani bütün insanlar bizim web sitemizin 4. adımında 20 saniye zaman geçirmiyor. Bunun üzerine gerekli kararları artık alabiliriz. Eğer bu H0'ı reddedemeseydik bu durumda kullanıcılarımızın web sitemizde bu 4. adımda yaklaşık 20 saniye zaman geçirdiğini kabul edecektik. Ve bu noktada buna göre bir iş kararı alabiliyor olacaktık. İşte bu şekilde tüm kullanıcı deneyimini hipotez testleri üzerine oturtturursak bu durumda elde edilecek olan kararlar tamamen bilimsel olur. Yani insan faktörü, yani bizim yorumlama yeteneklerimiz, bizim değerlendirme konusunda oluşturabileceğimiz yanlılıklar ortadan kaldırılmış oluyor böylece.

**Not:** Biz burada kolay olması açısından normallik varsayımını incelemedik. Bunun el yordamıyla teorik olarak incelenmesi biraz daha zor. Fakat uygulamalarımız kapsamında mutlaka netice itibariyle bu yapılacak olan hipotez testinin sonucuna gelmeden önce elimizdeki tek örneğin normallik varsayımını sağlayıp sağlamadığı test edilmelidir. Eğer normallik varsayımı sağlanmıyorsa non-parametrik t testi yapılmalıdır.
