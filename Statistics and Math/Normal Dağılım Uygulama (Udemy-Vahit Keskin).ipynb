## İş Uygulaması: Ürün Satış Olasılıklarının Hesaplanması

Ürün satış olasılıklarını normal dağılım aracılığıyla hesaplayacağız.

**Problem:**

Bir yatırım/toplantı öncesinde gelecek ay ile ilgili satışların belirli değerlerde gerçekleşmesi olasılıkları belirlenmek isteniyor. 

Bu toplantı şirketin yıllık toplantısı olabilir. Bazı departmanlara büyük çaplı belki yıllık, belki aylık hedefler verilecek olabilir. Ama ay tamamlanmadığından dolayı, ayın sonunu şu anda görmediğimizden dolayı (bilinmezlik olduğundan dolayı) bu bilinmezliği modellemeye çalışıyoruz. Bu her zaman bir toplantı ya da yatırım toplantısı öncesinde olacak diye bir durum söz konusu değil. Örneğin hiçbir koşula bağlı olmadan belirli sürekli değişkenlerin belirli değerleri alması olasılıklarıyla ilgileniyor olabiliriz. Örneğin bir futbolcunun 3 gol atması olasılığı, bir futbol takımının belirli sayıda gol atması olasılığı. Veya bir ekonomik göstergeyle ilgili bu göstergenin ortaya çıkması olasılığı gibi ekonomik göstergede de çok başarılı bir şekilde uyarlanabilmektedir. Örneğin Türkiye'nin şu göstergesinin şu değerden fazla olması olasılığı nedir gibi senaryolara da çok rahat bir şekilde uyum sağlayabilmektedir. 

Bu probleme dönecek olursak bize satış olasılıkları lazım bu satış olasılıkları üzerine bazı kararlar alacağız. 

**Detaylar:**

- Dağılımın normal olduğu biliniyor.
Nasıl bilinir bu? Görselleştirilir, görsel teknikler ile incelenir, normal dağıldığı gözüküyor olabilir veya testler yapılabilir (Hipotez testleri kısmında göreceğiz.)
- Aylık ortalama satış sayısı 80K, standart sapması 5K
- 90K'dan fazla satış yapma olasılığı nedir?

Zaten ortalamanın etrafında 75K ve 85K gibi değerlerde yer alınmasını beklerim. Sorumuz şu: 90K'dan fazla satış yapma olasılığı mı yüksektir? Yoksa 80K satış yapma olasılığı mı yüksektir? Zaten 75K ve 85K aralığının içinde veya bu aralığa yakın olan değerlerin olasılığının yüksek çıkmasını beklerim.   90K'dan fazla satış yapma olasılığının düşük olmasını bekleriz ama mesela 60K'dan fazla olması deseydik yüksek olmasını beklerdik. 

## Normal Dağılım

${\displaystyle f(x\mid \mu ,\sigma ^{2})={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}}}}$

* $\mu$  ortalama ya da dağılımın beklenen değeri

* ${\displaystyle \sigma }$  standart sapma

* ${\displaystyle \sigma ^{2}}$ varyans

İlgili değerleri fonksiyonda yerine yazdığımızda 90'dan fazla olma olasılığı 0.0228 çıkıyor.

70K'dan fazla olması olasılığını araştırsaydık olasılık değerinin arttığını gözlemliyoruz. Ortalama 80 olduğu için bunun daha fazla çıkması gerektiği ortadadır zaten.

73K'dan az olma olasılığına bakacak olursak: 

Satışların 85K ile 90K arasında olma olasılığı nedir diye bakacak olursak:

```python
from scipy.stats import norm
```

```python
# 90'dan fazla olması
1-norm.cdf(90,80,5)        

# Birinci argüman: hesaplamak istediğimiz değer
# İkinci argüman: ortalama
# Üçüncü argüman: standart sapma
```

cdf: cumulative density function (kümülatif yoğunluk fonksiyonu) kullanıyoruz çünkü kesiklilerden farklı olarak bu sefer örneğin 90'dan büyük bir alan taramış olacağız. Bu kümülatif bir yoğunluk olarak değerlendiriliyor.

90'dan büyük için hesaplama işlemi yaparken integral alma işlemi yaptığımız için 1'den çıkarma işlemi yaptık. Eğrinin altında kalan alanın tamamı 1'e eşit. 90'a kadar olan kısmı çıkartıyoruz yani.

```python
# 70'den fazla olması

1-norm.cdf(70,80,5) 
```

Ortalaması 80 olduğu için 70'den fazla olma olasılığının yüksek çıkması normal.

```python
# 73'den az olması

norm.cdf(73,80,5)  

# 1'den çıkarmamıza gerek yok.
```

```python
# 85 ile 90 arasında olması

norm.cdf(90,80,5)-norm.cdf(85,80,5)

# 90'a kadar olan kısımdan 85'e kadar olan kısmı çıkartıyoruz.
```
